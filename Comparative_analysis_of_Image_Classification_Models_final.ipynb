```python
#Data Preprocessing and creating batches of dataset
train_datagen = ImageDataGenerator(rescale=1./255,
                                   validation_split=0.2)

#Creating training Dataset
train_dataset = train_datagen.flow_from_directory(train_data_path,
                                                  target_size=(height, width),
                                                  batch_size=batch_size,
                                                  class_mode='categorical',
                                                  shuffle=True,
                                                  seed=seed,
                                                  color_mode='rgb',
                                                  interpolation='hamming',
                                                  subset='training')

```

```python
def model_accuracy(model_name):
  plt.plot(model_name.history['accuracy'])
  plt.plot(model_name.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()
```

```python
def model_loss(model_name):
  plt.plot(model_name.history['loss'])
  plt.plot(model_name.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()
```

## **Applying Covolutional Neural Network**

### **Model1 Using relu activation function**

```python
#Adding layers to CNN model
cnn_model = keras.models.Sequential([
    keras.layers.Conv2D(filters=150, kernel_size=(5,5), activation='relu', input_shape=(height,width,3)),# convolutional layer for feature extraction
    keras.layers.Conv2D(filters=200, kernel_size=(5,5), activation='relu'),
    keras.layers.Conv2D(filters=400, kernel_size=(5,5), activation='relu'),
    keras.layers.MaxPool2D(pool_size=(3, 3)),# maxpooling layer for reducing the dimentions(hight, width) of input tesor
    keras.layers.BatchNormalization(axis=-1),# normalize the activations of the previous layer

    keras.layers.Conv2D(filters=100, kernel_size=(3,3), activation='relu'),
    keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    keras.layers.MaxPool2D(pool_size=(2, 2)),
    keras.layers.BatchNormalization(axis=-1),
    keras.layers.Dropout(rate=0.25),#randmoly set i/p neuron to zero
    keras.layers.Flatten(),#multi-D to single-D
    keras.layers.Dense(512, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(rate=0.25),
    keras.layers.Dense(250, activation='relu'),

    keras.layers.Dense(43, activation='softmax')
])
```

```python
#Displaying Summary of neural network
cnn_model.summary()
```

```python
#compiling the CNN model

alpha=0.01
epochs=10
optim = keras.optimizers.Adam(learning_rate=0.01)
cnn_model.compile(optimizer = optim, loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
early = EarlyStopping(monitor='val_loss', patience=5, verbose=0, restore_best_weights=True)
```

```python
#fitting CNN model

trained_model_cnn = cnn_model.fit(train_dataset,
                    steps_per_epoch=len(train_dataset),
                    epochs=epochs,
                    validation_data=test_dataset,
                    validation_steps=len(test_dataset),
                    callbacks = early)
```

### **Implementing MobileNet Transfer Learning Model**

```python
train_dataset_Mv2 = train_datagen.flow_from_directory(train_data_path,
                                                  target_size=(224,224),
                                                  batch_size=batch_size,
                                                  class_mode='categorical',
                                                  subset='training')
```

```python
test_dataset_Mv2 = test_datagen.flow_from_directory(train_data_path,
                                                target_size=(224,224),
                                                batch_size=batch_size,
                                                class_mode='categorical',
                                                subset='validation')
```

```python
#Initialize mobileNetV2 model
image_size=[224,224]
mobileNet = MobileNetV2(input_shape=image_size+[3],weights='imagenet', include_top=False)
```

```python
#Freezing pre-trained model weights
for layer in mobileNet.layers:
    layer.trainable=False
```

```python
#Creating sequential model
mobilenet_model = Sequential()
mobilenet_model.add(mobileNet)
mobilenet_model.add(GlobalAveragePooling2D())
mobilenet_model.add(Dense(units=43, activation="softmax"))
```

```python
#Compiling MobileNetV2 model
mobilenet_model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001), loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
mobilenet_model.summary()
```

```python
model_check = ModelCheckpoint('best_model.h5', monitor='val_accuracy', verbose=0, save_best_only=True, mode='max')#Monitors validation_accuracy

early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=5, verbose=0, mode='max', restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)#Monitors validation_loss

csv_logger = CSVLogger('train_log.csv', separator=',')#Creates CSV file
```

```python
callbacks=[model_check, early, reduce_lr, csv_logger]
```

```python
print(len(train_dataset))
print(len(test_dataset))
```

```python
#Compiling MobileNetV2 model
alpha=0.001
optim = keras.optimizers.Adam(learning_rate=alpha)
mobilenet_model.compile(optimizer = optim, loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
mobilenet_history = mobilenet_model.fit(
  train_dataset_Mv2,
  validation_data=test_dataset_Mv2,
  epochs=10,
  steps_per_epoch=len(train_dataset_Mv2),
  validation_steps=len(test_dataset_Mv2),
  callbacks=callbacks
)
```

### **Implementing ResNet50 Transfer Learning Model**

```python
train_dataset_ResNet = train_datagen.flow_from_directory(train_data_path,
                                                  target_size=(224,224),
                                                  batch_size=batch_size,
                                                  class_mode='categorical',
                                                  subset='training')
```

```python
test_dataset_ResNet = test_datagen.flow_from_directory(train_data_path,
                                                target_size=(224,224),
                                                batch_size=batch_size,
                                                class_mode='categorical',
                                                subset='validation')
```

```python
# Load the ResNet50 model pretrained on ImageNet dataset
resnet_model = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the layers in the base ResNet50 model
for layer in resnet_model.layers:
    layer.trainable = False

# Build the classification head on top of the ResNet50 base
resnet_model = tf.keras.Sequential([
    resnet_model,
    layers.BatchNormalization(),
    layers.Flatten(),
    layers.Dense(350, activation='leaky_relu'),
    layers.Dense(43, activation='softmax')
])

# Print a summary of the model
resnet_model.summary()
```

```python
#Compiling ResNet50 Model
alpha=0.001
epochs=10
optim = keras.optimizers.RMSprop(learning_rate=alpha)
resnet_model.compile(optimizer = optim, loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
early = EarlyStopping(monitor='val_loss', patience=5, verbose=0, restore_best_weights=True)
```

```python
resnet_history = resnet_model.fit(train_dataset_ResNet,
                        steps_per_epoch=len(train_dataset_ResNet),
                        epochs=epochs,
                        validation_data=test_dataset_ResNet,
                        validation_steps=len(test_dataset_ResNet),
                        callbacks = early)
```

```python
resnet_model.save("resnet_model.h5")
```
